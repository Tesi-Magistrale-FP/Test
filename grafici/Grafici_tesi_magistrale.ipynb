# **Grafici dei test**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## **Grafico 1**

Scrittura di 100 transazioni semplici sul Tangle della Mainnet e della Devnet usando il PC.

```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
headers = ['Numero transazioni', 'Devnet', 'Mainnet']

df = pd.read_csv('https://raw.githubusercontent.com/Tesi-Magistrale-FP/Test_IOTA/main/grafici/G1_scrittura_semplice_pc_devnet_vs_mainnet.csv', names=headers)
df.set_index("Numero transazioni").plot(xlabel="Transazioni", ylabel="Latenza (ms)", title="Scrittura su Tangle - Devnet vs Mainnet")
plt.axhline(y = df['Devnet'].mean(), linestyle = ':')
plt.axhline(y = df['Mainnet'].mean(), color = 'orange', linestyle = ':')
plt.show()

print(df.describe())
```
